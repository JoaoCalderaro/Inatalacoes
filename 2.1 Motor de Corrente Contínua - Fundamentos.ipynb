3.	Uma fresadora demanda a mesma potência mecânica, independente da velocidade. Em um ponto de operação, mediu-se um torque de 30 N.m a uma rotação do motor de 200 rad/s. Nessa condição o motor de corrente contínua de imã permanente apresenta uma força contra-eletromotriz de 400 V, uma perda mecânica por atrito e ventilação de 600 W e uma resistência de armadura de 2,5 ohms. Considere que as perdas mecânicas variam linearmente com a velocidade dentro da faixa de operação. Despreze as perdas magnéticas e determine para a condição de carga de 10 N.m: a potência interna desenvolvida pelo motor, a corrente de armadura, a tensão de alimentação e o rendimento.

4. Nessa condição, o motor apresentou uma eficiência de 90% a uma velocidade de 100 rad/s.
